```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
df = pd.DataFrame(data=data.data , columns=data.feature_names)
```

```python
df['Target'] = data.target
```

```python
df
```

```python
# Because this implementation of Binary classification so that Why I am droping class 2
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[df['Target'] != 2]
```

```python
df.Target.unique()
```

```python
# Now defining X and y 

X = df.iloc[: , :-1]
y = df.iloc[: , -1]
```

```python
X
```

```python
y
```

```python
# Now Train test split 

from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X, y , test_size= 0.3 , random_state=1)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
# train model

from sklearn.linear_model import LogisticRegression

Model = LogisticRegression()
Model
```

```python
Model.fit(X_train , y_train)
```

```python
y_pred = Model.predict(X_test)
y_pred
```

```python
Model.predict_proba(X_test) #corresponding to higher probability class is predicted, cutoff is 0.5
```

```python
#evaluation metrics
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test , y_pred)
```

```python
accuracy_score(y_test , y_pred)
```

```python
print(classification_report(y_test , y_pred))
```

```python
#first two rows, 0 and 1 >> precision,  recall, f1 score and no of data points with respect to respective class
#macroavg >> for each class take simple avg>> avg precision of class 0 and class 1 will be macro average
#similarly for recall, f1 score
#support will be total no of datapoints

#weighted avg is used in case of class imbalance, it gives importance to performance of dominant class
```

```python
from sklearn.metrics import roc_curve , auc
```

```python
y_pred_proba = Model.predict_proba(X_test)[: ,1]
```

```python
y_pred_proba
```

```python
# ROC stand for Reciever Operating characterstics Curve 
#inputs : y_test, y_pred_proba. It will return TPR, FPR with different cutoff of probabilities
```

```python
fpr , tpr ,threshold = roc_curve(y_test , y_pred_proba)
```

```python
threshold
```

```python
tpr 
```

```python
fpr
```

```python
roc_auc = auc(fpr , tpr)
roc_auc
```

```python
plt.figure(figsize= (8,6))
plt.plot(fpr , tpr , color='darkorange' ,label= f'ROC curve(AUC={roc_auc})' )
plt.plot([0,1] , [0,1], color='red',linestyle='--' , label='Random guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```
