```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000 , n_features=20 , n_informative=5 , n_redundant=5 , n_classes=2 , random_state=42)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

Regressor = LogisticRegression()
Regressor
```

```python
Regressor.fit(X_train, y_train)
```

```python
y_pred = Regressor.predict(X_test)
y_pred
```

```python
y_pred_proba = Regressor.predict_proba(X_test)[: , 1]
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score  , classification_report

print(confusion_matrix(y_test , y_pred))
print(accuracy_score(y_test , y_pred))
print(classification_report(y_test , y_pred))
```

```python
from sklearn.metrics import roc_curve , auc 
```

```python
fpr , tpr , threshold = roc_curve(y_test , y_pred_proba)
```

```python
fpr
```

```python
roc_auc = auc(fpr , tpr)
roc_auc
```

```python
plt.figure(figsize=(8,6))
plt.plot(fpr , tpr , color='b' , lw=2,  label = f'ROC curve {roc_auc}')
plt.plot([0,1], [0,1] , color='r' ,lw=2, linestyle='--', label='Random Guess')
plt.xlabel('false positive rate')
plt.xlabel('true positive rate ')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```

```python
#by default prediction class is decided with cutoff 0.5
#but if you want to decide your own cutoff, you have to use precision, recall, accuracy tradeoff
#sensitivity specificity, accuracy tradeoff
```

```python
from sklearn.metrics import precision_score , accuracy_score , recall_score
```

```python
# Calculate precision, recall, and accuracy for different threshold probabilities
thresholds = np.linspace(0, 1, 100)
precisions = []
recalls = []
accuracies = []

for threshold in thresholds:
    y_pred_threshold = (y_pred_proba >= threshold).astype(int)
    precision = precision_score(y_test, y_pred_threshold)
    recall = recall_score(y_test, y_pred_threshold)
    accuracy = accuracy_score(y_test, y_pred_threshold)
    precisions.append(precision)
    recalls.append(recall)
    accuracies.append(accuracy)

# Plot precision, recall, and accuracy against threshold probabilities
plt.figure(figsize=(10, 6))
plt.plot(thresholds, precisions, label='Precision')
plt.plot(thresholds, recalls, label='Recall')
plt.plot(thresholds, accuracies, label='Accuracy')
plt.xlabel('Threshold Probability')
plt.ylabel('Score')
plt.title('Precision, Recall, and Accuracy vs. Threshold Probability')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#0.55 is my optimal threshold based on precision-recall-accuracy tradeoff
```

```python
new_pred_levels = np.where(y_pred_proba > 0.55 , 1 , 0)
```

```python
new_pred_levels
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, new_pred_levels)
```

```python
accuracy_score(y_test, new_pred_levels)
```

```python
print(classification_report(y_test, new_pred_levels))
```

```python
# Now we check accuracy through cross validation 

from sklearn.model_selection import KFold , cross_val_score

cv = KFold(n_splits=5 , shuffle=True )
```

```python
score = cross_val_score(Regressor , X_train , y_train , cv=cv)
score
```

```python
print(f"Model accuracy is {np.mean(score)}")
```
